## Kfold Cross Validation for Clasification

```python
import pandas as pd
from sklearn import model_selection
```

```python
df = pd.read_csv('./resources/wine_quality/winequality-red.csv')
df.head()
```

```python
df['kfold'] = -1
df.head()
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
kf = model_selection.KFold(n_splits=5)
```

```python
type(kf)
```

```python
for fold, (trn_, val_) in enumerate(kf.split(X=df)):
    df.loc[val_, 'kfold'] = fold
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df.head(40)
```

## Stratified Kfold Cross Validation for Regression

```python
import numpy as np
import pandas as pd

from sklearn import datasets
from sklearn import model_selection
```

```python
X, y = datasets.make_regression(
    n_samples=15000, n_features=100, n_targets=1
)
```

```python
print(X.shape)
print(y.shape)
```

```python
df = pd.DataFrame(
    X,
    columns = [f"f_{i}" for i in range(X.shape[1])]
)
```

```python
df['target'] = y
df.head().T
```

```python
df['kfold'] = -1

df.sample(frac=1).reset_index(drop=True)

df.head()
```

```python
num_bins = int(np.floor(1+np.log2(len(df))))
num_bins
```

```python
df['bins'] = pd.cut(
    df['target'], bins=num_bins, labels=False
)
df.head(10)
```

```python
kf = model_selection.StratifiedKFold(n_splits=5)
```

```python
for fold, (train, val) in enumerate(kf.split(X=df, y=df['bins'].to_numpy())):
    df.loc[val, 'kfold'] = fold
```

```python
df = df.drop("bins", axis=1)
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df.head()
```
